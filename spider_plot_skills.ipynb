```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
num_vars = 10

# Create a list of the skill levels
skills = [8, 7, 5, 8, 7, 4, 3, 6, 5, 4]

# Create a list of the skill names
labels = np.array(['Python', 'SQL', 'Knime/Zapier', 'Excel', 'Power BI',
                   'Dockers', 'Tensor Flow', 'SAP', 'GIT', 'Apache Spark'])

# Compute angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is made in a circular (not polygon) space, so we need to "complete the loop"
# and append the start to the end.
skills += skills[:1]
angles += angles[:1]

# Draw the plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, skills, color='red', alpha=0.25)
ax.plot(angles, skills, color='red', linewidth=2)  # Change the color if you like
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# Show the plot
plt.show()
```
